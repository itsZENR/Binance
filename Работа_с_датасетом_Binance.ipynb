```python
#@title Библиотеки
import pandas as pd
import numpy as np
import requests
import csv
import datetime
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
BINANCE_CSV_1h_ARBUSDT = "/content/binance_quotes_1h_ARBUSDT.csv"
BINANCE_CSV_1h_BELUSDT = "/content/binance_quotes_1h_BELUSDT.csv"
BINANCE_CSV_1h_BTCUSDT = "/content/binance_quotes_1h_BTCUSDT.csv"
BINANCE_CSV_1h_ETHUSDT = "/content/binance_quotes_1h_ETHUSDT.csv"
BINANCE_CSV_1h_XRPUSDT = "/content/binance_quotes_1h_XRPUSDT.csv"
```

```python
# Загружаем CSV-файл с данными и преобразуем данные в DataFrame
df1 = pd.read_csv(BINANCE_CSV_1h_ARBUSDT)[['open', 'close']]
df2 = pd.read_csv(BINANCE_CSV_1h_BELUSDT)[['open', 'close']]
df3 = pd.read_csv(BINANCE_CSV_1h_BTCUSDT)[['open', 'close']]
df4 = pd.read_csv(BINANCE_CSV_1h_ETHUSDT)[['open', 'close']]
df5 = pd.read_csv(BINANCE_CSV_1h_XRPUSDT)[['open', 'close']]
df1
```

```python
for j in [df1, df2, df3, df4, df5]:
  for i in range(1, 100):
    j[f'SMA{i}'] = j['close'].rolling(window=i).mean()
df1
```

```python
df1 = df1.iloc[:, 1:]
df2 = df2.iloc[:, 1:]
df3 = df3.iloc[:, 1:]
df4 = df4.iloc[:, 1:]
df5 = df5.iloc[:, 1:]
df1
```

```python
def SMA_test(df):
  for j in range(1, 100):
    predict = []
    target = []
    SMA = []
    procent = []
    for i in range(len(df)-1):
      # Если нынешняя цена меньше цены будущей
      if df.iloc[i, 0] < df.iloc[i+1, 0]:
        target.append(1)
      else:
        target.append(-1)
      
      if df.iloc[i, 0] < df.iloc[i, j]:
        predict.append(1)
      else:
        predict.append(-1)

    #  Считает скольок рах мы угадали
    sum = 0
    for i in range(len(target)):
      if target[i] == predict[i]:
        sum+=1
    print('на', df.iloc[:, j].name, 'всего', len(target), 'из них угадано', sum)
    SMA.append(df.iloc[:, j].name)
    print('%', sum/len(target)*100)
    procent.append(sum/len(target)*100)
    print()
```

```python
for i, j in enumerate([df1, df2, df3, df4, df5]):
  print('номер датасета', i+1)
  SMA_test(j)
  print('-'*100)
```
